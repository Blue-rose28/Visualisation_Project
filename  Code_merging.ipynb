## Import Capital Health Expanditure data and Hygiene Data

```python
#create path to csv file:
gdp_path= Path('Resources/gdp_per_capita.csv')
unemployment_path= Path('Resources/unemployment.csv')
literacy_path= Path('Resources/literacy.csv')
health_expenditure_path= Path('Resources\health_expenditure.csv')

#read csv files: 

gdp_df=pd.read_csv(gdp_path)
unemployment_df=pd.read_csv(unemployment_path)
literacy_df=pd.read_csv(literacy_path)
health_expenditure_df=pd.read_csv(health_expenditure_path)
```

```python
#check unemployment file
gdp_df.head()
```

Merge smoking and gdp per capita

```python
# Merge the DataFrames
smoking_gdp = smoking_df.merge(gdp_df, on=['Country Name', 'Year'], how='left')
smoking_gdp.head()
```

```python
# Drop the 'Country Code' column from the DataFrame
smoking_gdp.drop(columns=['Country Code'], inplace=True)
smoking_gdp.head()
```

```python
#check unemployment file
unemployment_df.head()
```

Merge smoking, GDP per capita and Unemployment

```python
smoking_gdp_unemployment = smoking_gdp.merge(unemployment_df, on=['Country Name', 'Year'], how='left')

smoking_gdp_unemployment.drop(columns=['Country Code'], inplace=True)

smoking_gdp_unemployment
```

 Merge smoking, GDP per capita and Unemployment, Literacy

```python
smoking_gdp_unemployment_literacy = smoking_gdp_unemployment.merge(literacy_df, on=['Country Name', 'Year'], how='left')

smoking_gdp_unemployment_literacy.drop(columns=['Country Code'], inplace=True)

smoking_gdp_unemployment_literacy
```

 Merge smoking, GDP per capita and Unemployment, Literacy, health expenditure create merge_all file

```python
merge_all = smoking_gdp_unemployment_literacy.merge(health_expenditure_df, on=['Country Name', 'Year'], how='left')

merge_all.drop(columns=['Country Code'], inplace=True)

merge_all
```

```python
#Export to csv file namely Master_raw
output_file_path= Path('Resources\ master_raw.csv') 
merge_all.to_csv(output_file_path, index=False)
```

```python
#Drop missing value in merged file
master = merge_all.dropna()
master
```
